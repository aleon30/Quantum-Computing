```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(1,1)
qc.reset(0)
qc.x(0)
qc.measure(0, 0)
qc.draw('mpl')
```

```python
qc_x = QuantumCircuit(1)
qc_x.x(0)
qc_x.draw('mpl')
```

```python
from qiskit.quantum_info import Operator, Statevector
X = Operator(qc_x)
X.draw('latex')
```

```python
ket_0 = Statevector.from_label('0')
ket_0.draw('latex')
```

```python
ket_0 = ket_0.evolve(X)
ket_0.draw('latex')
```

```python
ket_1 = Statevector.from_label('1')
ket_1.draw('latex')
```

```python
ket_1 = ket_1.evolve(X)
ket_1.draw('latex')
```
